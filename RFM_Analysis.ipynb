```python
def monetary(d):
    if d['TOTALREVENUE']>=5000:
        return '1.VIP'
    elif d['TOTALREVENUE']>=1000:
        return '2.HIGH'
    elif d['TOTALREVENUE']>=500:
        return '3.MEDIUM'
    elif d['TOTALREVENUE']>=100:
        return '4.LOW'
    elif d['TOTALREVENUE']<100:
        return '5.VERY LOW'
mont['Monetary Flag']=mont.apply(monetary,axis=1)
```

```python
def recency(year):
    if year == 2019 or year==2020 or year==2021:
        return 'Purchased in Before 2 Years'
    elif year == 2022:
        return 'Purchased in last Year'
    elif year == 2023:
        return 'Purchase in this Year'
    else:
        return 'Other'  # Provide a default value for unmatched cases
```

```python
rec['Recency Flag']=rec['OrderYear'].apply(recency)
```
